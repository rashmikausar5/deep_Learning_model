# **Image Agumentation**

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
```

```python
# load Data

img = load_img('/content/drive/MyDrive/women.webp')     # showing pic
```

```python
x = img_to_array(img)
print(x.shape)
```

```python
# reshape dimension to 4

x = x.reshape((1, )+x.shape)
print(x.shape)
```

```python
# variable defination

datagen = ImageDataGenerator( rotation_range = 40,
                             width_shift_range = 0.2,
                             height_shift_range =0.2,
                             rescale = 1./255,
                             shear_range = 0.2,
                             zoom_range = 0.2,
                             horizontal_flip = True,
                             vertical_flip = True,
                             fill_mode = 'nearest'
                             )
```

```python
# create synthetic image

i = 0
for batch in datagen.flow(x, batch_size=1, save_to_dir = '/content/drive/MyDrive/Datasets/Output' , save_prefix='women_syc', save_format='jpeg'):
    i +=1
    if i > 20:
      break # otherwise loop infinetly
```

```python
# create sythetic image

i = 0
for batch in datagen.flow(x, batch_size=1,
                          save_to_dir='/content/drive/MyDrive/Colab Notebooks',
                          save_prefix='women_syn', save_format='jpeg'):
    i += 1
    if i > 20:
        break  # otherwise the generator would loop indefinitely
```
